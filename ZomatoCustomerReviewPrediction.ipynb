```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


dataset = pd.read_excel('Dataset/ZomatoCustomerReviewPrediction.xlsx')
dataset.head()

```

```python
dataset.shape
```

```python
city=dataset.iloc[:,2]
output=dataset.iloc[:,1]
c1=0
c2=0
c3=0
c4=0
c5=0
c6=0
```

```python
for i in range(0,1999):
    if(city[i]=="hyderabad" and output[i]==1):
        c1=c1+1
    if(city[i]=="hyderabad"and output[i]==0):
        c2=c2+1
    if(city[i]=="bangalore" and output[i]==1):
        c3=c3+1
    if(city[i]=="bangalore"and output[i]==0):
        c4=c4+1
    if(city[i]=="delhi" and output[i]==1):
        c5=c5+1 
    if(city[i]=="delhi" and output[i]==0):
        c6=c6+1        
```

```python
objects = ('Hyderabad', 'Bangalore', 'Delhi')
y_pos = np.arange(len(objects))
pos = [c1,c3,c5]
neg =[c2,c4,c6]
bw=0.2
plt.bar(y_pos, pos, width=0.2,color='green', label='Positive Review')
plt.bar(y_pos+bw, neg,  width=0.2,color='red', label='Negative Review')
plt.xticks(y_pos, objects)
plt.ylabel('Feedback')
plt.title('Zomato Customer Review')
plt.legend()
plt.show()
```

```python
labels = 'Positive', 'Negative'
sizes = [(c1+c3+c5),(c2+c4+c6)]
colors = ['gold', 'lightcoral']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.legend()
plt.show()
```

```python
import re
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
data=[]
for i in range(0,2000):
         review=dataset["Review"][i]
         review=re.sub("[^a-zA-Z]"," ",review)
         review=review.lower()
         review=review.split()
         review=[ps.stem(word) for word in review if not word in set (stopwords.words("english"))]
         review= " ".join(review)
         data.append(review)
```
